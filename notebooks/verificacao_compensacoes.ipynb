```python
import pandas as pd

# Lista de arquivos CSV
arquivos = [
    r"P:\Common\IQS\CIS\IQS_VIOLACOES_DFM_2026-06_C.csv",
    r"P:\Common\IQS\CIS\IQS_VIOLACOES_DFM_2026-06_L.csv",
    r"P:\Common\IQS\CIS\IQS_VIOLACOES_DFM_2026-06_M.csv",
    r"P:\Common\IQS\CIS\IQS_VIOLACOES_DFM_2026-06_P.csv",
    r"P:\Common\IQS\CIS\IQS_VIOLACOES_DFM_2026-06_V.csv"
]

# Carregar o primeiro arquivo com o cabeçalho
df_principal = pd.read_csv(arquivos[0], encoding='UTF-8',sep = ';',decimal=',')

# Carregar os demais arquivos ignorando o cabeçalho
dfs_restantes = [pd.read_csv(arq, encoding='UTF-8',sep = ';',decimal=',', header=None, skiprows=1) for arq in arquivos[1:]]

# Ajustar os cabeçalhos dos arquivos restantes para os do primeiro arquivo
for df in dfs_restantes:
    df.columns = df_principal.columns

# Concatenar todos os DataFrames
df_final = pd.concat([df_principal] + dfs_restantes, ignore_index=True)

df = df_final
# Agrupando por SIGLA_RDIS 
result = df.groupby('SIGLA_REG').agg({ 'MAIOR_COMP_PREV': 'sum', 'NUM_UC_VINS': pd.Series.nunique }).reset_index() 
# Renomeando as colunas para um nome mais claro 
#result.columns = ['SIGLA_RDIS', 'total_COMP_PREV', 'distinct_NUM_UC_VINS']

# Adicionando a linha de totalizador
total_row = pd.DataFrame({
    'SIGLA_REG': ['Total'],
    'MAIOR_COMP_PREV': [result['MAIOR_COMP_PREV'].sum()],
    'NUM_UC_VINS': [result['NUM_UC_VINS'].sum()]
})

result = pd.concat([result, total_row], ignore_index=True)

print(result)
```
